# 1점

```python
url = 'https://www.tripadvisor.com/Attraction_Review-g294197-d1169465-Reviews-N_Seoul_Tower-Seoul.html'

driver = webdriver.Chrome()
driver.get(url)
time.sleep(2)

driver.find_element(By.XPATH, '/html/body/div[1]/main/div[1]/div[2]/div[2]/div[2]/div/div[1]/section[9]/div/div/div/section/section/div[1]/div/div[1]/div/div/div[2]/div/div/div[1]/div/button').send_keys(Keys.ENTER)
time.sleep(1)
driver.find_element(By.XPATH, '/html/body/div[5]/div/div[2]/div/div[2]/div/button[1]').send_keys(Keys.ENTER)
driver.find_element(By.XPATH, '/html/body/div[5]/div/div[3]/div/div[1]/button').send_keys(Keys.ENTER)
time.sleep(1)

data_1 = []
j = 1
while(j < 7):
#for j in range(2, 4):
    j += 1
    for i in range(2, 12):

        try:
            if j == 2:
                review = driver.find_element(By.XPATH, f'/html/body/div[1]/main/div[1]/div[2]/div[2]/div[2]/div/div[1]/section[9]/div/div/div/section/section/div[1]/div/div[5]/div/div[{i}]/div/div/div[5]/div[1]/div').text
                data_1.append(review)
                time.sleep(1)
            else:
                review = driver.find_element(By.XPATH, f'/html/body/div[1]/main/div[1]/div[2]/div[2]/div[2]/div/div[1]/section[9]/div/div/div/section/section/div[1]/div/div[5]/div/div[{i}]/div/div/div[5]/div[1]/div/span').text
                data_1.append(review)
                time.sleep(1)
        except:
            continue

    try:
        driver.find_element(By.XPATH, '/html/body/div[1]/main/div[1]/div[2]/div[2]/div[2]/div/div[1]/section[9]/div/div/div/section/section/div[1]/div/div[5]/div/div[11]/div[1]/div/div[1]/div[2]/div/a').send_keys(Keys.ENTER)
        time.sleep(2)
    except:
        continue

    
    
driver.close()

```

```python
pd.DataFrame(data_1).to_csv('seoul_tower_1.csv')
```
